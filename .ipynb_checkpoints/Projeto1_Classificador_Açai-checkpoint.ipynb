### samldnask - LIMPEZA

```python
import functools
import operator
import re
import emoji
```

```python
texto = str(df.Treinamento.sum()); texto

replace_list = ["https://",".",":",",","''",'"',"#","-","$","\n","\n\n","_","?","!","/",";","{","}","|","(",")","__","___","\\",
                "¬¥","`","^","~","¬∫","¬™","*","&","¬®","%","@","¬£","¬¢","¬¨","  ","   ","    ","     "]
def Replace_all(texto):
    for r in replace_list:
        texto = texto.replace(r," ")
    std = re.compile('[!\-.:?;]'); texto = re.sub(std," ",texto)
    return texto

def Splitter(texto):
    return functools.reduce(operator.concat, [substr.split() for substr in emoji.get_emoji_regexp().split( texto )])

texto = Replace_all(texto.lower()); texto = Splitter(texto); texto = pd.Series(texto)
print("Palavras do texto:"); pd.DataFrame(texto).T
```

```python
classif_0 = str(df[df.A==0].Treinamento.sum()); classif_0
classif_0 = Replace_all(classif_0.lower()); classif_0 = Splitter(classif_0); classif_0 = pd.Series(classif_0)
print("Palavras da classif 0:"); pd.DataFrame(classif_0).T
```

```python
classif_1 = str(df[df.A==1].Treinamento.sum()); classif_1
classif_1 = Replace_all(classif_1.lower()); classif_1 = Splitter(classif_1); classif_1 = pd.Series(classif_1)
print("Palavras da classif 1:"); pd.DataFrame(classif_1).T
```

___
### Montando um Classificador Naive-Bayes - TREINAMENTO

```python
dados_teste = pd.read_excel('a√ßai2.xlsx', sheet_name = 'Teste')
acai_idx = dados_teste.set_index('Teste'); acai_idx
acai_idx_teste = dados_teste.set_index('A')
tweets_teste = acai_idx_teste.values.tolist()
```

```python
dados_teste;
```

```python
texto = texto.value_counts(True)

freq_0 = classif_0.value_counts(True)
freq_1 = classif_1.value_counts(True)

abs_0 = classif_0.value_counts(False)
abs_1 = classif_1.value_counts(False)
```

```python
Probabilidade_relevantes = len(freq_1)/len(texto)
Probabilidade_irrelevantes = len(freq_0)/len(texto)
```

```python
Probabilidade_relevantes, Probabilidade_irrelevantes
```

```python
# Fun√ß√£o que classifica os tweets para a base de teste
def classification(tweet,a,b):
    if a>b:
        return 1
    else:
        return 0 
Lista_classif_0_final = []
Lista_classif_1_final = []
Classification = []

for tweet in tweets_teste:
    for frase in tweet:
        lista_classif_0 = []
        lista_classif_1 = []
        frase_limpa = Replace_all(frase.lower()) ; frase_limpa = Splitter(frase)
        palavras = frase_limpa
        
         # Laplace   
        prob_relevante = (abs_1[palavras].fillna(0) + 1)/(len(set(texto)) + len(classif_1))
        prob_irrelevante = (abs_0[palavras].fillna(0) +1)/(len(set(texto)) + len(classif_0))
        lista_classif_1.append(prob_relevante)
        lista_classif_0.append(prob_irrelevante)
        Lista_classif_1_final.append(lista_classif_1)
        Lista_classif_0_final.append(lista_classif_0)
        
        # Naive Bayes
        for frase in Lista_classif_1_final:
            for palavra in frase:
                Prob_rel = palavra.prod() *Probabilidade_relevantes   
        for frase in Lista_classif_0_final:
            for palavra in frase:
                Prob_irrel = palavra.prod() * Probabilidade_irrelevantes
        Classific = classification(tweet, Prob_rel,Prob_irrel) 
        Classification.append(Classific)
        
```

___
### Verificando a performance do Classificador - TESTE

Verificando a performace do classificador no produto escolhido para testar o classificador

```python
dados_teste["Simula√ß√£o"] = Classification
dados_teste[:50]

# Tabela com a Simula√ß√£o feita pelo classificador
```

```python
pos = 0
falso_pos = 0
neg = 0
falso_neg = 0
total = len(dados_teste["A"])

for [e,i] in zip(dados_teste["A"], dados_teste["Simula√ß√£o"]):
    if e == 1:
        if i == 1:
            pos += 1
        else: 
            falso_neg += 1
    if e == 0:            
        if i == 0:
            neg +=1
        else:
            falso_pos +=1            
```

___
### Concluindo

```python
print("Verdadeiros relevantes: ", round((pos/total)*100,2), "%")
print("Falsos relevantes: ", round((falso_pos/total)*100, 2), "%")
print("Verdadeiros irrelevantes: ", round((neg/total)*100,2), "%")
print(f"Falsos irrelevantes: ", round((falso_neg/total)*100, 2), "%")
print("_"*35)

print("Acertos: ", round(((pos/total)+(neg/total))*100,2), "%")
print("Erros: ", round(((falso_pos/total)+(falso_neg/total))*100,2), "%")
```
